```python
import pandas as pd
  

def createSeries (series_list):
    
  # create a series
  series_list = pd.Series(series_list)
    
  return series_list
  
# create a series of s1
s1 = createSeries([3.000000, 2.666667, 2.333333, 2.00000])  
# create a series of s2
s2 = createSeries(['3', '2', '1'])
# create a series of s3
s3 = createSeries([30568.630973,
         8048.278464,26203.877303,4451.139508,13920.197502])
# create a dictonary
data = {"s1": s1,
        "s2": s2,
        "s3": s3}
  

df = pd.concat(data,
               axis = 1)
  
# show the dataframe
df
```

# step 9

```python
def stats(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = newData.apply(stats)
loc_stats
```

# step 10

```python
day_stats = newData.apply(stats,axis=1)
day_stats.head()
```
